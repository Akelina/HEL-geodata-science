Przeprowadź klasteryzację danych iris, na ile pokrywa się ona z oryginałem.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans,SpectralClustering,DBSCAN
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df=pd.read_csv('iris_dane.csv',sep=',')
df.head()
```

```python
df['label'].unique()
```

```python
X=np.array(df.iloc[:,[0,1,2,3]])
Y=np.array(df.iloc[:,[4]])
```

```python
kmeans_iris = KMeans(n_clusters=3)
#kmeans_iris=SpectralClustering(n_clusters=3)
#kmeans_iris=DBSCAN(min_samples=5)
```

```python
kmeans_iris.fit(X)
```

```python
XL=kmeans_iris.labels_
```

```python
print(XL)
```

```python
Y
```

```python
YL=Y.reshape(-1)
```

```python
YL
```

```python
print(confusion_matrix(YL,XL))
```

```python
print(classification_report(YL,XL))
```

### t-SNE

 t-Distributed Stochastic Neighbor Embedding (t-SNE) służy do wizualizacji położenia w przestrzeni multispektralnej i jest dimensionality reduction technique

```python
from sklearn.manifold import TSNE
```

```python
model = TSNE(learning_rate=100)
```

```python
transformed = model.fit_transform(X)
```

```python
# Plotting 2d t-Sne
x_axis = transformed[:, 0]
y_axis = transformed[:, 1]


f, ax = plt.subplots()
for i in np.unique(YL):
    mask = XL == i
    plt.scatter(x_axis*mask, y_axis*mask,label=i)
ax.legend();
```

Zadanie 2

Zastosuj metodę t-SNE do wizualizacji położenia w przestrzeni 4000 elementów ze zbioru  MNIST

```python


## Load the training set
train_data = np.load('MNIST/mnist_train_data.npy')
train_labels = np.load('MNIST/mnist_train_labels.npy')

train_dataB=train_data[:4000,:]
train_labelsB=train_labels[:4000]
```

```python
len(train_dataB)
```

```python
model = TSNE(learning_rate=200)
```

```python
transformed = model.fit_transform(train_dataB)
```

```python
x_axis = transformed[:, 0]
y_axis = transformed[:, 1]


f, ax = plt.subplots(figsize=(10,10))

for i in np.unique(train_labelsB):
    mask = train_labelsB == i
    plt.scatter(x_axis*mask, y_axis*mask,label=i)
ax.legend();
```
