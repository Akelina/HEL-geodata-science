### Modes

Są dwa sposoby działania komórek (modes)  **edit**  i **command**. Naciśnij ESC aby przejść do **command** i ENTER aby przejść do **edit**.

**Edit** służy do pisania tekstu i kodu w komórce. Ramka komórki ma kolor <font color='green'>ZIELONY</font>.
<br>
**Command** służy do wykonywania komend edycyjnych takich jak wycinanie, wklejanie, wstawianie komórek.Ramka komórki ma kolor <font color='blue'>NIEBIESKI</font>.
