### Concatenating, Merging, and Joining DataFrames

**Concatenating**<br>
Concatenation basically glues DataFrames together. When concatenating DataFrames, keep in mind that dimensions should match along the axis you are concatenating on. 


```python
d1=pd.DataFrame(np.random.rand(5,4),range(0,5),['A','B','ID','C'])
```

```python
d2=pd.DataFrame(np.random.rand(5,4),range(0,5),['D','E','ID','F'])
```

```python
pd.concat([d1,d2],axis=0)
```

```python
pd.concat([d1,d2],axis=1)
```

**Merging**

```python
for i in range (0,5): 
    d1.loc[i,'ID']=i
print(d1)    
for i in range (0,5): 
    d2.loc[i,'ID']=i
print(d2)    
    
```

```python
dff=pd.merge(d1,d2,how='outer',on='ID')
dff
```

dff.sort_values('ID',ascending=False)

**Join**

```python
for i in range (0,5): 
    d1.loc[i,'ID']=i
print(d1)    
for i in range (0,5): 
    d2.loc[i,'IDD']=i
print(d2)  
```

```python
dff2=d1.set_index('ID').join(d2.set_index('IDD'))
dff2
```
