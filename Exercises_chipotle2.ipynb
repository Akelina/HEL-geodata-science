### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo=pd.read_csv('chipotle.tsv',sep='\t')
chipo.head(30)
```

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
chipo.head(3)
```

### Step 4. Jaka jest cena każdego item_name? 
###### podaj item_name i  cenę

```python
grupy=chipo.groupby(by="item_name") 
pp=grupy[['item_price']].min()
print(pp)
```

### Step 6. Przesortuj tabelę od najwyższej ceny

```python
pp.sort_values('item_price',ascending=False)
```

### Step 7. Jaką zapłacono największą cenę za zamówienie?

```python
grupy=chipo.groupby(by="order_id") 
pp=grupy[['item_price']].sum()
#print(pp)
pp.describe()
```

### Step 8. Jak wiele razy w zamówieniu pojawiła się  Veggie Salad Bowl ?

```python
chipo_veg=chipo[chipo['item_name']=='Veggie Salad Bowl']
len((chipo_veg))  #*chipo_veg['quantity'])
```
