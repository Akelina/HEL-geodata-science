
**Matplotlib** jest pakietem (zestawem modułów Pythona) do tworzenia dwuwymiarowych wykresów o jakości publikacyjnej w różnych formatach i środowiskach interaktywnych (IPython, jupyter notebook).  Opis biblioteki jest zawarty na stronie https://matplotlib.org/

Interfejsem dostępy tego pakietu jest **pyplot**. Pracę z pakietem zaczynamy praktycznie zawsze linią kodu:

```python
import matplotlib.pyplot as plt
```

```python
from numpy import *
from random import *

yy=zeros((10),float)
xx=zeros((10),float)
for i in range(0,10):
    yy[i]=uniform(0,10)
    xx[i]=i

plt.plot(xx,yy);
```

 Legenda, tytuł, opis osi

```python


x=[1,2,3]
y=[3,7,4]

plt.plot(x,y)
plt.xlabel('Zmienne')
plt.ylabel('Nasza funkcja')
plt.title('Wykres\nPrzykład 1');
```

```python
x=[1,2,3]
y=[3,7,4]

x1=[1,2,3]
y1=[10,11,8]

plt.plot(x,y,label='Pierwsza linia')
plt.plot(x1,y1,label='Druga linia')
plt.xlabel('Zmienne')
plt.ylabel('Nasza funkcja')
plt.title('Wykres\nPrzykład 1a')
plt.legend();
```

<p style="font-family: Arial; font-size:1.75em;color:purple; font-style:bold"><br>
 Wykres słupkowy i histogram

```python


x=[2,4,6,8,10]
y=[6,7,8,2,4]

plt.bar(x,y,label='Bars1')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph')
plt.legend();



```

```python


x=[2,4,6,8,10]
y=[6,7,8,2,4]

x2=[1,3,5,9,11]
y2=[7,8,2,4,2]

plt.bar(x,y,label='Bars1')
plt.bar(x2,y2,label='Bars2')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph')
plt.legend();
```
