## Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
midwest=pd.read_csv('midwest_filter.csv')
```

```python
midwest.head()
```

```python
midwest.columns
```

### Histogramy 

```python
plt.subplots(figsize=(6,6))
sns.distplot(midwest['popwhite']);
```

```python
plt.subplots(figsize=(6,6))
sns.distplot(midwest['popwhite'],bins=15,kde=False,rug=True);
```

```python
plt.subplots(figsize=(6,6))
sns.distplot(midwest['popwhite'],bins=15,hist=False,rug=True);
```

```python
plt.subplots(figsize=(6,6))
sns.kdeplot(midwest['popwhite'], shade=True);
```

```python
plt.subplots(figsize=(8,8))
sns.kdeplot(midwest['popamerindian'], shade=True);
sns.kdeplot(midwest['popasian'], shade=True);
sns.kdeplot(midwest['popblack'], shade=True);
```

### Scatterplots

```python
plt.subplots(figsize=(10,10))
sns.scatterplot(x='area',y='poptotal',data=midwest);
plt.xlabel('Powierzchnia')
plt.ylabel('Populacja');
```

```python
plt.subplots(figsize=(10,10))
sns.scatterplot(x='area',y='poptotal',data=midwest,hue='category');
plt.xlabel('Powierzchnia')
plt.ylabel('Populacja');
```

```python
plt.subplots(figsize=(10,10))
sns.scatterplot(x='area',y='poptotal',data=midwest,hue='popwhite',size='popwhite',sizes=(10,200));
plt.xlabel('Powierzchnia')
plt.ylabel('Populacja');
```

```python
plt.subplots(figsize=(10,10))
sns.regplot(x='area',y='poptotal',data=midwest);
```

```python
sns.jointplot(x='area',y='poptotal',data=midwest,height=7);
```

```python
sns.jointplot(midwest.area,midwest.poptotal,kind="hex",color="r",height=10);
```
